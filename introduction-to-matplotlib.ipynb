## Matplotlib Example Workflow

```python
# 0. Import matplot lib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plots
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (save whole figure)
fig.savefig("complete/images/sample-figure.png")
```

## Making Figures with NumPy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data to create scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
butter_prices = {"Almond butter":10,
                 "Peanut butter":8,
                 "Cashew butter":12}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Nut Butter Prices",
       ylabel="Price($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
```

```python
# Historgram next. Make some data.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(10000))
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(butter_prices.keys(), butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```
